```python
stroke2=stroke.drop(['id','gender','Residence_type','work_type','ever_married','smoking_status'],axis=1)
stroke2
```

```python
stroke2.to_csv('stroke2.csv')
pd.read_csv('stroke2.csv')
```

```python
sto=stroke2.values
sto
```

```python
x=sto[0:,0:5]
x
```

```python
y=sto[:,5]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,shuffle=True,random_state=42)
print("x_train:",x_train.shape)
print("y_train:",y_train.shape)
print("x_test:",x_test.shape)
print("y_test:",y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.svm import SVC
```

```python
logmodel=LogisticRegression(class_weight='balanced')
logmodel.fit(x_train,y_train)
```

```python
#age,hypertension,heart_disease,avg_gulcose_level,,bmi
prediction=logmodel.predict([[49,1,0,100,36]])
prediction
```

```python
prediction=logmodel.predict([[67,0,1,228.69,36.6]])
print(prediction)
prediction=logmodel.predict([[40,0,0,100.34,28.1]])
print(prediction)
```

```python
prediction1=logmodel.predict(x_test)
prediction1.shape
```

```python
y_test
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
accuracy_score(y_test,prediction1)
```

```python
confusion1=confusion_matrix(y_test,prediction1)
print(confusion1)
```

* 3 actual positives incorrectly predicted as negatives
* 114 actual negatives incorrectly predicted as positives

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(confusion1,cmap="jet",annot=True)
plt.show()
```

```python
classification_report(y_test,prediction1)
```
